Use the data saved at ```./data/company_sales_data.csv``` to use for all exercises.

### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1

import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
# company_sales.describe()
company_sales
````

````python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
````

````python
fig = plt.figure()
ax = plt.axes()

ax.plot(company_sales.month_number, company_sales.total_profit)
ax.set(xlim=(1, 12), ylim=(150000, 450000),
       xlabel='month number', ylabel='total profit',
       title='Profit Data By Month');
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
# fig = plt.figure()
ax = plt.axes()

ax.plot(company_sales.month_number, company_sales.total_profit, ':or', linewidth=3, label='total profit amt by month')
ax.set(xlim=(1, 12), ylim=(150000, 450000),
       xlabel='month number', ylabel='sold units number',
       title='Profit Data By Month');
plt.legend();
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# Exercise 3
ax = plt.axes()
plt.plot(company_sales.month_number, company_sales.facecream, '-ok', label='facecream')
plt.plot(company_sales.month_number, company_sales.facewash, '--vb', label='facewash')
plt.plot(company_sales.month_number, company_sales.toothpaste, '-.pg', label='toothpaste')
plt.plot(company_sales.month_number, company_sales.bathingsoap, ':*c', label='bathing soap')
plt.plot(company_sales.month_number, company_sales.shampoo, '--dm', label='shampoo')
plt.plot(company_sales.month_number, company_sales.moisturizer, ':+r', label='moisturizer')
plt.legend(loc='upper left')
ax.set(xlim=(1, 12), ylim=(0, 15000),
       xlabel='month number', ylabel='sold units number',
       title='Monthly Sold Units of Each Item');
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
# Exercise 4
ax = plt.axes()
plt.scatter(company_sales.month_number, company_sales.toothpaste, marker='.', color="purple")
ax.grid(linestyle='-')
ax.set(xlim=(1, 12), ylim=(4000, 9000),
       xlabel='month number', ylabel='sold units of toothpaste',
       title='Monthly Sold Units of Toothpaste');
````
